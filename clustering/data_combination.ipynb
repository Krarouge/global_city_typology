#### Roads cleanup

```python
columns_to_drop = ['bus_guideway', 'cycleway', 'busway', 'bus_route', 
                   'construction', 'yes', 'planned', 'bridleway', 'proposed', 
                   'platform', 'rest_area']
roads_global_cleaned = roads_global.drop(columns=columns_to_drop)

# drop unnecessary columns
columns_to_drop = ['bus_guideway', 'cycleway', 'busway', 'bus_route', 
                   'construction', 'yes', 'planned', 'bridleway', 'proposed', 
                   'platform', 'rest_area']
roads_global_cleaned = roads_global.drop(columns=columns_to_drop)



```

```python
# Define the columns to be summed
columns_to_sum_motorway = ['motorway', 'motorway_link']
columns_to_sum_asphalt = ['trunk', 'trunk_link', 'primary', 'primary_link', 'secondary', 'secondary_link', 'tertiary', 'tertiary_link', 'residential', 'residential_link', 'living_street', 'military', 'raceway', 'service', 'services']
columns_to_sum_others = ['road', 'unclassified', 'track']

# Replace commas with dots and convert to numeric
roads_global_cleaned[columns_to_sum_motorway] = roads_global_cleaned[columns_to_sum_motorway].replace({',': '.'}, regex=True).apply(pd.to_numeric, errors='coerce')
roads_global_cleaned[columns_to_sum_asphalt] = roads_global_cleaned[columns_to_sum_asphalt].replace({',': '.'}, regex=True).apply(pd.to_numeric, errors='coerce')
roads_global_cleaned[columns_to_sum_others] = roads_global_cleaned[columns_to_sum_others].replace({',': '.'}, regex=True).apply(pd.to_numeric, errors='coerce')

# Fill NaN values with 0
roads_global_cleaned[columns_to_sum_motorway] = roads_global_cleaned[columns_to_sum_motorway].fillna(0)
roads_global_cleaned[columns_to_sum_asphalt] = roads_global_cleaned[columns_to_sum_asphalt].fillna(0)
roads_global_cleaned[columns_to_sum_others] = roads_global_cleaned[columns_to_sum_others].fillna(0)

```

```python




# sum up the specified columns into new ones
roads_global_cleaned['road_motorway'] = roads_global_cleaned[columns_to_sum_motorway].sum(axis=1)
roads_global_cleaned['road_asphalt'] = roads_global_cleaned[columns_to_sum_asphalt].sum(axis=1)
roads_global_cleaned['road_others'] = roads_global_cleaned[columns_to_sum_others].sum(axis=1)

# Drop the original individual columns that were summed
columns_to_drop_summed = columns_to_sum_motorway + columns_to_sum_asphalt + columns_to_sum_others
roads_global_cleaned = roads_global_cleaned.drop(columns=columns_to_drop_summed)

# Disp first few rows of cleaned dataframe
roads_global_cleaned.head()
```

```python
# Create the 'ID' column
urban_co2_emissions['ID'] = urban_co2_emissions.index + 1
```

Distance to sea

```python
gdf_dist_sea["distance_meters"] = gdf_dist_sea["distance"] * 111320 
```

Data gathering

```python
# Merging the dataframes on their respective ID 
merged_df = urban_co2_emissions.merge(roads_global_cleaned[['ID', 'road_motorway', 'road_asphalt', 'road_others']], 
                                      on='ID', how='inner')

merged_df = merged_df.merge(ghs_stat_ucdb[['ID_HDC_G0', 'AREA', 'P15', 'P00', 'E_GR_AV14', 'B15', 'GDP15_SM', 'GDP00_SM', 'EX_HW_IDX']], 
                            left_on='ID', right_on='ID_HDC_G0', how='inner')

merged_df.rename(columns={
    'ID_HDC_G0_x': 'ID_HDC_G0'
    }, inplace=True)

# Merge with gdf_dist_sea
merged_df = merged_df.merge(gdf_dist_sea[['ID_HDC_G0', 'distance_meters']], 
                            on='ID_HDC_G0', how='left')

# Renaming columns
merged_df.rename(columns={
    'ID_HDC_G0_x': 'ID_HDC_G0',
    'P15': 'pop_2015', 
    'P00': 'pop_2000', 
    'E_GR_AV14': 'green_2014', 
    'B15': 'built_2015', 
    'GDP15_SM': 'GDP_2015', 
    'GDP00_SM': 'GDP_2000', 
    'EX_HW_IDX': 'heatwave'
}, inplace=True)

# Computing pop_growth
merged_df['pop_growth'] = ((merged_df['pop_2015'] / merged_df['pop_2000']) ** (1/15) - 1) * 100

# Compute GDP_growth
merged_df['GDP_growth'] = ((merged_df['GDP_2015'] / merged_df['GDP_2000']) ** (1/15) - 1) * 100

# drop the columns that are no longer needed
final_columns = ['ID_HDC_G0','CTR_MN_NM', 'UC_NM_MN', 'AREA', 'road_motorway', 'road_asphalt', 'road_others', '2022', 
                 'Mean_Growth_Rate', 'Mean_Growth_2018_2022', 'pop_2015', 'pop_growth', 'green_2014', 'built_2015', 
                 'GDP_2015', 'GDP_growth', 'heatwave', 'distance_meters']

final_df = merged_df[final_columns]  # Drop pop_2000 after computation

# Display the first few rows of the final dataframe
final_df.head()
```
